```python
# Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Mengatur tampilan pandas agar lebih rapi
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.width', 1000)       # Menyesuaikan lebar tampilan

# Ganti nama file dengan path yang sesuai jika perlu
file_name = 'datauts.csv'

# Pastikan file ada sebelum membaca
if os.path.exists(file_name):
    # Membaca dataset
    data = pd.read_csv(file_name)

    print("\n10 Data Teratas dari Dataset:")
    print(data.head(10))  # Menampilkan 10 baris pertama dengan pengaturan baru

    # e. Apa produk yang paling banyak dibeli?
    if 'Product line' in data.columns:
        product_counts = data['Product line'].value_counts()
        most_bought_product = product_counts.idxmax()

        print(f"\nE. Produk yang paling banyak dibeli: {most_bought_product} (Jumlah: {product_counts[most_bought_product]})")
        
        # Cetak langsung total penjualan masing-masing produk
        print("\nTotal Penjualan Masing-Masing Produk:")
        print(product_counts)

        # Visualisasi 6 Produk Terbanyak Dibeli
        product_counts.head(6).plot(kind='bar', color='purple', title='6 Produk Terbanyak Dibeli')
        plt.xlabel('Produk')
        plt.ylabel('Jumlah Terjual')
        plt.title('Produk Terbanyak Dibeli')
        plt.show()

        # Tambahan Visualisasi 1: Total Penjualan per Kota
        if 'City' in data.columns:
            city_counts = data['City'].value_counts()
            print("\nTotal Penjualan Masing-Masing Kota:")
            print(city_counts)

            city_counts.plot(kind='bar', color='green', figsize=(10, 6), title='Total Penjualan per Kota')
            plt.xlabel('Kota')
            plt.ylabel('Jumlah Terjual')
            plt.title('Distribusi Penjualan Berdasarkan Kota')
            plt.show()

            # Tambahan Visualisasi 2: Penjualan per Produk dan Kota (Heatmap)
            product_city_pivot = data.pivot_table(index='Product line', columns='City', aggfunc='size', fill_value=0)
            print("\nPenjualan per Produk dan Kota:")
            print(product_city_pivot)

            plt.figure(figsize=(12, 8))
            sns.heatmap(product_city_pivot, annot=True, fmt="d", cmap="Blues")
            plt.title("Penjualan per Produk dan Kota")
            plt.xlabel("Kota")
            plt.ylabel("Produk")
            plt.show()
    else:
        print("\nKolom 'Product line' tidak ditemukan.")
else:
    print(f"\nFile '{file_name}' tidak ditemukan. Pastikan path file benar.")
```

```python
# Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Mengatur tampilan pandas agar lebih rapi
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.width', 1000)       # Menyesuaikan lebar tampilan

# Ganti nama file dengan path yang sesuai jika perlu
file_name = 'Dp.csv'

# Pastikan file ada sebelum membaca
if os.path.exists(file_name):
    # Membaca dataset
    data = pd.read_csv(file_name)

    print("\n10 Data Teratas dari Dataset:")
    print(data.head(10))  # Menampilkan 10 baris pertama dengan pengaturan baru

    # Mengubah data ke dalam format yang lebih mudah dipahami untuk visualisasi
    pendidikan_columns = [
        'Tidak/belum pernah sekolah', 'Tidak/belum tamat SD', 'SD', 'SLTP',
        'SLTA Umum/SMU', 'SLTA Kejuruan/SMK', 'Akademi/Diploma', 'Universitas'
    ]
    data_melted = data.melt(id_vars=['Periode', 'Bulan'], value_vars=pendidikan_columns, var_name='Pendidikan', value_name='Jumlah Pengangguran')

    # 1. Ringkasan Pengangguran Berdasarkan Pendidikan
    pengangguran_total = data_melted.groupby('Pendidikan')['Jumlah Pengangguran'].sum().reset_index()
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(data=pengangguran_total, x='Pendidikan', y='Jumlah Pengangguran')
    plt.xlabel('Pendidikan')
    plt.ylabel('Jumlah Pengangguran')
    plt.title('Total Pengangguran Berdasarkan Pendidikan')
    plt.xticks(rotation=45)

    # Menambahkan label di atas setiap bar
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'),
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center', 
                    xytext = (0, 10), 
                    textcoords = 'offset points')

    plt.show()

    # 2. Tren Pengangguran dari Waktu ke Waktu
    plt.figure(figsize=(12, 8))
    sns.lineplot(data=data_melted, x='Periode', y='Jumlah Pengangguran', hue='Pendidikan', marker='o')
    plt.xlabel('Periode')
    plt.ylabel('Jumlah Pengangguran')
    plt.title('Tren Pengangguran Berdasarkan Pendidikan dari Waktu ke Waktu')
    plt.xticks(rotation=45)
    plt.legend(title='Pendidikan')
    plt.show()

    # 3. Perbandingan Pengangguran di Bulan Tertentu

    print("""
          
                                                         Data Khusus Februari
          
          """)

    selected_month = 'Februari'  # Ganti dengan bulan yang ingin dibandingkan
    data_february = data_melted[data_melted['Bulan'] == selected_month]
    plt.figure(figsize=(12, 8))
    ax = sns.barplot(data=data_february, x='Periode', y='Jumlah Pengangguran', hue='Pendidikan', palette='coolwarm')
    plt.xlabel('Periode')
    plt.ylabel('Jumlah Pengangguran')
    plt.title(f'Perbandingan Pengangguran di Bulan {selected_month}')
    plt.xticks(rotation=45)
    plt.legend(title='Pendidikan')

    # Menambahkan label di atas setiap bar
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'),
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center', 
                    xytext = (0, 10), 
                    textcoords = 'offset points')

    plt.show()
    print("""
          
                                                         Data Khusus Agustus
          
          """)
    selected_month = 'Agustus'  # Ganti dengan bulan yang ingin dibandingkan
    data_february = data_melted[data_melted['Bulan'] == selected_month]
    plt.figure(figsize=(12, 8))
    ax = sns.barplot(data=data_february, x='Periode', y='Jumlah Pengangguran', hue='Pendidikan', palette='coolwarm')
    plt.xlabel('Periode')
    plt.ylabel('Jumlah Pengangguran')
    plt.title(f'Perbandingan Pengangguran di Bulan {selected_month}')
    plt.xticks(rotation=45)
    plt.legend(title='Pendidikan')

    # Menambahkan label di atas setiap bar
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'),
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center', 
                    xytext = (0, 10), 
                    textcoords = 'offset points')

    plt.show()
else:
    print(f"\nFile '{file_name}' tidak ditemukan. Pastikan path file benar.")
```
